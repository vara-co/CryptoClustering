```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
line_plot = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)

line_plot
```

To view the above plot in the github repository, visit the "images" directory

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_scaled = StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", 
                                                                  "price_change_percentage_7d", 
                                                                  "price_change_percentage_14d",
                                                                  "price_change_percentage_30d",
                                                                  "price_change_percentage_60d",
                                                                  "price_change_percentage_200d",
                                                                  "price_change_percentage_1y"]])

# Show the first 8 rows of scaled data
market_scaled[0:8]
```

```python
# Create a DataFrame with the scaled data
market_scaled_df = pd.DataFrame(market_scaled, columns = ["price_change_percentage_24h", 
                                                           "price_change_percentage_7d", 
                                                           "price_change_percentage_14d",
                                                           "price_change_percentage_30d",
                                                           "price_change_percentage_60d",
                                                           "price_change_percentage_200d",
                                                           "price_change_percentage_1y"])

# Copy the crypto names from the original data
market_scaled_df["coin_id"] = df_market_data.index

# Set the coinid column as index
market_scaled_df = market_scaled_df.set_index("coin_id")

# Display sample data
market_scaled_df.head(8)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for value in k:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=value)
# 2. Fit the model to the data using `df_market_data_scaled`
    k_model.fit(market_scaled_df)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Show DataFrame
elbow_df.head(8)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k
)

elbow_plot
```

To view the above plot in the github repository, visit the "images" directory

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for k according to the above graphic is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
crypto_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
crypto_model.fit(market_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = crypto_model.predict(market_scaled_df)

# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the DataFrame
MarketScaled_copy = market_scaled_df.copy()

# Show copy df
MarketScaled_copy.head(8)
```

```python
# Add a new column to the DataFrame with the predicted clusters
MarketScaled_copy["market_predictions"] = k_4

# Display sample data
MarketScaled_copy.head(8)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot =MarketScaled_copy.hvplot.scatter(x="price_change_percentage_24h",
                                 y="price_change_percentage_7d",
                                 c="market_predictions",
                                 cmap="colorblind",  
                                 hover_cols=["coin_id"],
                                )

scatter_plot

# I used the colors that best suited this cluster plot to actually be able to see the different clusters https://holoviews.org/user_guide/Colormaps.html
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.

marketdata_pca = pca.fit_transform(MarketScaled_copy) 

# View the first five rows of the DataFrame. 
marketdata_pca[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# Sum the explained variance of the pca and get percentage to answer the following question
ttl_explained_variance = sum(explained_variance) * 100

# Show both results
print("The explained variance is: ", explained_variance)

print("The answer to the following question is: ", ttl_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is about 89%

```python
# Create a new DataFrame with the PCA data.
crypto_market_df = pd.DataFrame(
    marketdata_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)

# Copy the crypto names from the original data
crypto_market_df["coin_id"]=df_market_data.index

# Set the coinid column as index
crypto_market_df=crypto_market_df.set_index("coin_id")

# Display sample data
crypto_market_df.head(8)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=0)
# 2. Fit the model to the data using `df_market_data_pca`
    k_model.fit(crypto_market_df)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)

# Show a sample of the DataFrame
pca_elbow_df.head(8)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_plot = pca_elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title="PCA Elbow Curve",
    xticks=k
)

pca_elbow_plot
```

To view the above plot in the github repository, visit the "images" directory

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(crypto_market_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
kpca_4 = model.predict(crypto_market_df)

# Print the resulting array of cluster values.
kpca_4
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** In this scenario, 2 of the clusters only have 1 datapoint each. Thus having less features doesn't seem to impact the overall interpretation of the cluster, as in both cases they are very similar. Using fewer features might be helpful when the dataset is heavy with data that might not be relevant to what we are trying to convey. However, it might also have a negative impact if we end up removing what we think is unnecessary, and ends up being crucial information at the end. I suppose it has it's pros and cons. Maybe if you want to do a quick clustering for analysis it would be a good way to go. However, if you want to go deeper and finding other insights, it might be better to use the original data.
  * Fewer features can simplify computations with an easier interpretation of results, which is advantageous if processing complex amounts of data. It can also improve the clustering process.
  * The down side is the potential for loss of valuable information that could be relevant for clustering. In this particular case, there wasn't much difference, but this is because we are seeing both options and comparing. If we were choosing to just do the PCA version, we wouldn't know if it's introducing bias or not.
